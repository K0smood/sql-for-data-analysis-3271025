```python
#creando variables para usar a lo largo del proceso
a単adir_estudiante = True
notavalida = True
estudiantes = []
i = 1
l = 1

#Ingreso de estudiantes
while a単adir_estudiante == True:
    nombre = input("Ingrese el nombre del estudiante: ")
    while nombre == "" or nombre.isdigit(): #verifica que se haya ingresado un nombre y que no sea un numero
          nombre = input("Por favor ingrese un nombre valido ")
    apellido = input("Ingrese el apellido del estudiante: ")
    while apellido == "" or apellido.isdigit(): #verifica que se haya ingresado un apellido y que no sea un numero
          apellido = input("Por favor ingrese un apellido valido ")

    nombre_completo = nombre + " " + apellido #combina nombre y apellido
    estudiantes.append(nombre_completo) #guarda el nombre completo en un espacio de la lista

    respuesta = input("Desea ingresar otro estudiante? ")
    while respuesta != "Si" and respuesta != "No": #solicita explicitamente un "Si" o "No" como respuesta
        respuesta = input("Por favor responda con 'Si' o 'No': ")
    if respuesta == "Si":
        a単adir_estudiante = True
    elif respuesta == "No":
        a単adir_estudiante = False


cantidad = len(estudiantes) #creando variable con cantidad de estudiantes ingresados
promedios = [] #creando variable que contendra todos los promedios al finalizar

#Ingreso de notas
for i in range(cantidad): #inicia un bucle que se repetira segun la cantidad de estudiantes que fueron ingresados

    nota1 = input("Por favor ingrese la primer nota de " + estudiantes[i] + ", debe ser un numero del 1 al 10: ") #se aprovecha la variable i para tomar el valor correspondiente de la lista de estudiantes
    if nota1.isdigit() and 1 <= int(nota1) <= 10: #verifica que se haya ingresado un numero entero entre 1 y 10
        notavalida = True
    else:
        notavalida = False
        while notavalida == False:
            nota1 = input("Por favor ingrese una nota valida (un numero entero del 1 al 10): ")
            if nota1.isdigit() and 1 <= int(nota1) <= 10: #verifica que se haya ingresado un numero entero entre 1 y 10
                notavalida = True
            else:
                notavalida = False

    nota2 = input("Por favor ingrese la segunda nota de " + estudiantes[i] + ", debe ser un numero del 1 al 10: ")
    if nota2.isdigit() and 1 <= int(nota2) <= 10:
        notavalida = True
    else:
        notavalida = False
        while notavalida == False:
            nota2 = input("Por favor ingrese una nota valida (un numero entero del 1 al 10): ")
            if nota2.isdigit() and 1 <= int(nota2) <= 10:
                notavalida = True
            else:
                notavalida = False

    nota3 = input("Por favor ingrese la tercer nota de " + estudiantes[i] + ", debe ser un numero del 1 al 10: ")
    if nota3.isdigit() and 1 <= int(nota3) <= 10:
        notavalida = True
    else:
        notavalida = False
        while notavalida == False:
            nota3 = input("Por favor ingrese una nota valida (un numero entero del 1 al 10): ")
            if nota3.isdigit() and 1 <= int(nota3) <= 10:
                notavalida = True
            else:
                notavalida = False

    #se convierten las notas a valor de numero entero y se calcula el promedio
    nota1 = int(nota1)
    nota2 = int(nota2)
    nota3 = int(nota3)
    promedio = (nota1 + nota2 + nota3) / 3
    promedio = round(promedio, 2)

    #guarda el promedio correspondiente en un espacio de la lista
    promedios.append(promedio)

    i += 1

#Resumen final
for l in range(cantidad): #inicia un bucle que se repetira segun la cantidad de estudiantes que fueron ingresados
  if promedios[l] >= 6: #verifica si el promedio que esta siendo usado en el bucle es mayor o igual a 6 para determinar si mostrar un mensaje de aprobado o desaprobado
    print(estudiantes[l] + " ha aprobado con un promedio de: " + str(promedios[l]))
  else:
    print(estudiantes[l] + " ha reprobado con un promedio de: " + str(promedios[l]))

  l += 1
```
